```python
def handlerate(value): 
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value [0]
        return float(value)
df_zomato['rate']=df_zomato['rate'].apply(handlerate) 
df_zomato['rate'].head()
```

```python
def removecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df_zomato['cost2plate']=df_zomato['cost2plate'].apply(removecomma)
```

```python
df_zomato['cost2plate'].unique()
```

```python
rest_types=df_zomato['rest_type'].value_counts(ascending=False) 
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]  #restaurant type having less than 1000 shops
```

```python
rest_types_lessthan1000
```

```python
#making restaurant in others category having less than 1000
```

```python
def having_less(value):
    if value in rest_types_lessthan1000:
        return 'others'
    else:
        return value
df_zomato['rest_type']=df_zomato['rest_type'].apply(having_less)
```
